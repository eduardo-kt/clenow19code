# Execução do "simply python simulation"

## Importar bibliotecas

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import yfinance as yf
```

## Obter dados

```python
tick = yf.Ticker('vale3.sa')
data = tick.history(period = '5y', interval = '1d', auto_adjust = True)
```

## Calcular médias

```python
data['MA50'] = data['Close'].rolling(50).mean()
data['MA100'] = data['Close'].rolling(100).mean()
```

## Obter sinais

```python
data['signal'] = np.where(data['MA50']>data['MA100'],1,0)

#delay the trade to the day after the signal
data['signal'] = data['signal'].shift() 
```

## Obter percentuais da estratégia

```python

data['strategy_pct'] = data['Close'].pct_change(1) * data['signal']
```

## Calcular desempenho da estratégia 

```python

data['strategy'] = (data['strategy_pct'] + 1).cumprod()
data['buyhold'] = (data['Close'].pct_change(1) + 1).cumprod()
```

## Plotar estratégias

```python
fig = go.Figure()
fig.add_scatter(x = data.index, y = data['buyhold'], name= 'Vale3')
fig.add_scatter(x = data.index, y = data['strategy'], name= 'Strategy')

fig.update_layout(template = "plotly_white")
fig.show()
```
