# Correlação com subplots

# Bibliotecas

```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

## Importar dados

```python
ticks = yf.Tickers('engi11.sa wege3.sa')
data = ticks.history(period = '5y', interval = '1d')
```

## Separar preços de fechamento e calcular retornos diários

```python
closeprices = data.loc[:,'Close'].pct_change()
```

## Calcular a correlação entre os ativos 
* Para uma determinada janela
* Para um determinado número de dias (padrão 250 dias == 1 ano)

```python
window = 50
dias = 250

correlation = closeprices['ENGI11.SA'].rolling(window).corr(closeprices['WEGE3.SA'])[-dias:]
```

## Nivelar dados
* criar um nível para os ativos de maneira que seja possível comparar os desempenhos


```python
dias = 250
performance_new_base = (closeprices[-dias:] + 1).cumprod()
```

## Relative Strength

```python
dias = 250
relative_s = data[('Close','ENGI11.SA')]/data[('Close','WEGE3.SA')]
relative_s = relative_s.iloc[-dias:]
```

## Subplots

* Comparação retorno dos ativos
* Relative Strength
* Correlação

```python
fig = make_subplots(rows=3,cols=1, subplot_titles=("Relative Performance", "Relative Strength", "Correlation"))
fig.add_scatter(x = performance_new_base.index, y = performance_new_base['ENGI11.SA'], row=1, col=1, name='ENGIE')
fig.add_scatter(x = performance_new_base.index, y = performance_new_base['WEGE3.SA'], row=1, col=1, name= "WEGE")

fig.add_scatter(x = relative_s.index, y = relative_s, row=2, col=1, showlegend=False)

fig.add_scatter(x = correlation.index, y = correlation, row=3, col=1, showlegend= False)

fig.update_layout(template = 'plotly_white')
fig.show()
```
